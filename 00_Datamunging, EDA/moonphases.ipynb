```python
moon_phase_list = []
for i in moon_map['ts_date']:
    lowest_difference_between_dates = 0
    for index, date_moon_tuple in enumerate(zip(moons['datetime'], moons['phase'])):

        try:
            a = abs(pd.to_datetime(date_moon_tuple[0])-pd.to_datetime(i))
            b = date_moon_tuple[1]

            if index == 0:
                lowest_difference_between_dates = a
            else:
                if a < lowest_difference_between_dates:
                    lowest_difference_between_dates = a
                    lowest_phase = b
        except:
            continue
    moon_phase_list.append((i, lowest_phase))
```
